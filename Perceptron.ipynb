```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()

# features and labels separation (inputs and outputs)
X = iris.data
# Iris has 3 types of flowers; we will do a binary classification for simplicity (hence below we will use only two classes)
y = (iris.target != 0).astype(int)

# all rows, 1st column
sepal_length = X[:, 0]
# all rows, 2nd column
sepal_width = X[:, 1]

plt.figure(figsize=(8, 6))
plt.scatter(sepal_length[y == 0], sepal_width[y == 0], color='blue', marker='o', label='Setosa (class 0)')
plt.scatter(sepal_length[y == 1], sepal_width[y == 1], color='red', marker='x', label='Non-Setosa (class 1)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset: Setosa vs Non-Setosa - Linear Separability')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from Perceptron import Perceptron
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
X = iris.data
y = (iris.target != 0).astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

perceptron = Perceptron(num_inputs=4, learning_rate=0.01, epochs=10)
perceptron.train(X_train, y_train)

predictions = [perceptron.predict(inputs) for inputs in X_test]

accuracy = np.mean(predictions == y_test)

print(f"Accuracy on test dataset: {accuracy:.2f}")
```
